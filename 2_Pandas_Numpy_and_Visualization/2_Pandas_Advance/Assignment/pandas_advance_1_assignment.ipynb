### Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
df = pd.read_csv(r"./dataset/data.csv")
def re_indexing(df):
    df = df.reset_index(drop=True)
    df.index = pd.RangeIndex(start=1, stop=2*len(df), step=2)
    return df
```

### Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
df = pd.read_csv(r"./dataset/text.csv")
def create_word_count(df):
    # create a new column 'Word_Count'
    df['Word_Count'] = df['Text'].apply(lambda x: len(x.split()))
    return df

df
```

### Q5. How are DataFrame.size() and DataFrame.shape() different?

**Ans:** The difference between DataFrame.size() and DataFrame.shape() are:

* **DataFrame.size():** This method returns the number of elements in the DataFrame, which is equivalent to the total number of cells in the table. It is calculated as the product of the number of rows and the number of columns.
<br><br>
* **DataFrame.shape():** This attribute returns a tuple representing the dimensions of the DataFrame in the form (rows, columns). Therefore, it provides the number of rows and the number of columns in the DataFrame.

```python
df = pd.read_csv(r'./dataset/text.csv')
df
```

```python
print(f"DataFrame.size() = {df.size}\n\nDataFrame.shape() = {df.shape}\n")
```

### Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

```python
df = pd.read_csv(r'./dataset/email.csv')

def extract_username(df):
    df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])
    return df
df
```

### Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows. 

### For example, if df contains the following values:

| A | B | C |
| :-: | - | :-: |
| 3 | 5 | 1 |
| 8 | 2 | 7 |
| 6 | 9 | 4 |
| 2 | 3 | 5 |
| 9 | 1 | 2 |

### Your function should select the following rows: 
|   A  |   B  |   C  |
| ---- | ---- | ---- |
|   8  |   2  |  7   |
|   9  |   1  |  2   |

The function should return a new DataFrame that contains only the selected rows.

```python
df = pd.read_csv(r'./dataset/data.csv')
df = df.drop('Values', axis=1)

def select_rows(df):
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10) & (df['C'] != 4)]
    return selected_rows

df
```

```python
select_rows(df)
```

### Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.

```python
df = pd.read_csv(r'./dataset/data.csv')

def mean_(df):
    
    return (sum(df['Values']) / len(df['Values']))

def median_(df):
    sorted_values = sorted(df['Values'])
    n = len(sorted_values)
    if n % 2 == 0:
        median = (sorted_values[n//2 + 1] + sorted_values[n//2]) / 2
    else:
        median = sorted_values[n//2]
    
    return median
        
        
def std_dev(df):
    mean = sum(df['Values']) / len(df['Values'])
    variance = sum((x - mean) ** 2 for x in df['Values']) / (len(df['Values']) - 1)
    std_dv = variance ** 0.5
    return std_dv

df
```

```python
median_(df)
```

```python
std_dev(df)
```
