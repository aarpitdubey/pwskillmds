## Python Variable 
- A Python variable can be thought of as a "labeled" (named) box or a bucket where you can keep different kinds of information. 
- In Python, you can use a variable to store numbers, words, and other types of data

 ---------------------
|      Variable       |
|---------------------|
|       Data          |

---------------------

- In this diagram, the "Variable" is like a box with a label on it, and inside the box, you can put different types of "Data"

- When we write below python code 
```Python
    Variable = "Data"
```
Python internally allocate some memory and store "Data" which is referred by a variable named as Variable.




````python
Variable = "Data"
````
