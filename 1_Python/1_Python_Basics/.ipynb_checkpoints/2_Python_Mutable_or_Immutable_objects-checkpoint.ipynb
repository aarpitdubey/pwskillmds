There is another data type in python that is List datatype. It is denoted by ``l = []`` (pair of square brackets) or list() function (l = list())

```python
l = [2, 3, "Appu", 3 + 7j, True, 34.56]
```

```python
type(l)
```

A list in Python is a collection of items that are ordered and changeable. Each item in the list has an assigned index, which represents its position in the list.

 -----------------------------------
| Index 0 | Index 1 | Index 2 | ... |
|---------|---------|---------|-----|
|  Item   |  Item   |  Item   | ... |
 -----------------------------------
 
 here, ``l = [2, 3, "Appu", 3 + 7j, True, 34.56]``
 
 -----------------------------------
| Index: |    0    |     1   |    2    |    3   |  4  | 5   | 
|-----|---------|---------|---------|--------|-----|-----|
| Data: |    2    |    3    |  "Appu" | 3 + 7j |True |34.56|
-----------------------------------
| Index: |   -6    |     -5   |    -4    |    -3   |  -2  | -1   |
|---------|---------|---------|---------|--------|-----|-----|
| Data: |    2    |    3    |  "Appu" | 3 + 7j |True |34.56|
 -----------------------------------

In Python, mutability refers to whether an object's state can be modified after it has been created. An object is mutable if its state (the value it holds) can be changed after it is created. Conversely, an object is immutable if its state cannot be changed after it is created.

Lists in Python are mutable, which means that you can change the elements of a list after it has been created. You can add, remove, or modify the items in a list.

Strings in Python are immutable, which means that once a string is created, you cannot change its individual characters. If you want to modify a string, you need to create a new string.

**Note:** lists are mutable because you can change their elements after creation, while strings are immutable because you cannot change their individual characters once the string is created.
